```python
#ваш код здесь
def get_experience(value: str):
    if pd.isnull(value):
        return np.nan
    if value.lower() == 'не указано':
        return np.nan
    exp = value.split(' ')[2:]
    if (exp[1].startswith('мес')):
        return int(exp[0])
    months = 12 * int(exp[0])
    if len(exp[2]) > 0 and exp[2].isdigit():
        months += int(exp[2])
    return months
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(get_experience)
df['Опыт работы (месяц)'].median()
df.drop(df[['Опыт работы']], axis=1, inplace=True)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
def parse_mobility(s: str):
    def get_city(city: str):
        city = city.strip()
        if (city in ['Москва', 'Санкт-Петербург']):
            return city
        if city in million_cities:
            return 'город-миллионник'
        return 'другие'
    def is_ready_to_trip(readiness: str):
        if readiness.strip().startswith('не '):
            return False
        return True
    spltted = s.split(' ,')
    city = get_city(spltted[0])
    index = 1
    if spltted[index].strip().startswith('м.'):
        index +=1
    is_ready_to_relocation = is_ready_to_trip(spltted[index])
    if(index +1 == len(spltted)):
        is_ready_to_work_trip = False
    else:
        is_ready_to_work_trip = is_ready_to_trip(spltted[index+1])
    return [city, is_ready_to_relocation, is_ready_to_work_trip]
def split_mobility(serie: pd.Series):
    res = list()
    for i in serie:
        res.append(parse_mobility(str(i)))
    result = np.array(res)
    return pd.Series(result[:, 0]), pd.Series(result[:, 1]), pd.Series(result[:, 2])
        


df['Город'], df['Готовность к переезду'], df['Готовность к командировкам'] = split_mobility(df['Город, переезд, командировки'])
len(df[( df['Готовность к командировкам'] == 'True') & (df['Готовность к переезду'] == 'True')]) / len(df) *100

df.drop(df[['Город, переезд, командировки']], axis=1, inplace=True)
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
exchange_rates = pd.read_csv('ExchangeRates.csv')
exchange_rates.tail(100)

def split_salary(salary: str):
    splitted = salary.split(' ')
    currency = splitted[1].strip('.')
    if currency == 'бел.руб':
        return [int(splitted[0]), 'BYN']
    if currency == 'грн':
        return [int(splitted[0]), 'UAH']
    if currency == 'сум':
        return [int(splitted[0]), 'UZS']
    if len(currency[1].split('.')) > 1:
        currency = ''.join(currency.split('.'))
    return [int(splitted[0]), currency]
def get_salary_expectations(serie: pd.Series):
    res = list()
    for i in serie:
        res.append(split_salary(i))
    result = np.array(res)
    return pd.Series(result[:, 0]), pd.Series(result[:, 1])
df['Обновление резюме'] = df['Обновление резюме'].apply(lambda x : pd.to_datetime(x, dayfirst=True))
df['зарплата'], df['валюта'] = get_salary_expectations(df['ЗП'])
exchange_rates['date'] = exchange_rates['date'].apply(lambda x : pd.to_datetime(x, dayfirst=True))
df['валюта'].unique()
merged = (pd.merge(df, exchange_rates, how='left', left_on='валюта', right_on='currency'))
merged[merged['валюта'] != 'руб']
```

```python
result_data = merged[(merged['Обновление резюме'].dt.date == merged['date'].dt.date) | (merged['валюта'] == 'руб')]
result_data['close'].fillna(1.0, inplace=True)
result_data['proportion'].fillna(1.0, inplace=True)
result_data['ЗП в рублях'] = (result_data['close'] / result_data['proportion']) * (1.0 * result_data['зарплата'].apply(lambda x: float(x))) 
#result_data[result_data['валюта'] != 'руб']
#result_data['ЗП в рублях'].median()
result_data.info()
```
